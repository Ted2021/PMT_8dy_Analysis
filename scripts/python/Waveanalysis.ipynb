```python
import uproot
import matplotlib.pyplot as plt
```

```python
tree = uproot.open("/Volumes/cta/work/dora_data/pmt/230503/ZQ7568/20230503_ZQ7568_source.root:Treesingle_0")
tree.show()

time = tree.arrays(["time"], library="np")["time"]
wform1 = tree.arrays(["wform1"], library="np")["wform1"]
wform0 = tree.arrays(["wform0"], library="np")["wform0"]
wform3 = tree.arrays(["wform3"], library="np")["wform3"]
wform2 = tree.arrays(["wform2"], library="np")["wform2"]
```

```python
stopcell = tree.arrays(["stopcell"], library="np")["stopcell"]
```

```python
def plot_show(i):
    plt.figure(figsize=(20, 6))
    plt.subplot(1,4,1)
    plt.plot(time[i], wform0[i], color="blue", label = str(i))
    plt.axvline(x = time[i][391], linestyle = "--", color = "black")
    plt.title("wform0")
    plt.legend()
    
    plt.subplot(1,4,2)
    plt.plot(time[i], wform1[i], color="orange", label = str(i))
    plt.axvline(x = time[i][391], linestyle = "--", color = "black")
    plt.title("wform1")
    plt.legend()
    
    plt.subplot(1,4,3)
    plt.plot(time[i], wform2[i], color="lightgreen", label = str(i))
    plt.axvline(x = time[i][391], linestyle = "--", color = "black")
    plt.xlim(time[i][391-30],time[i][391+30])
    plt.title("wform2")
    plt.legend()
    
    plt.subplot(1,4,4)
    plt.plot(time[i], wform3[i], color="red", label = str(i))
    plt.axvline(x = time[i][391], linestyle = "--", color = "black")
    #plt.xlim(time[i][391-30],time[i][391+30])
    plt.title("wform3")
    plt.legend()
    
    print(stopcell[i])
```

```python
plot_show(1)
```

```python
for j in range(100):
    for i in range(1024):
        if (i+stopcell[j]%1024==392):
            print(str(j) + " : " + str(i))
```
